2. Create the below pattern using nested for loop in Python.




```python
#code to print the pattern

n=int(input("Enter n : "))
for i in range(n+1):
    for j in range(i):
        print("*",end="")
    print()
for i in range(n,0,-1):
    for j in range(j):
        print("*",end="")
    print()
```
